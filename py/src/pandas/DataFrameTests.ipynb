# Apply: General Split-Apply-Combine

```python
def top(df, n=5, column='tip_pct'):
    """Selects the rows with the largest values in
        a particular column"""
    return df.sort_values(by=column)[-n:]
```

```python
top(tips, n=6)
```

```python
tips.groupby('smoker').apply(top)
```

```python
# specify function parameters
tips.groupby(['smoker', 'day']).apply(top, n=1, column='total_bill')
```

```python
# Avoiding Multiindex
tips.groupby('smoker', group_keys=False).apply(top)
```

### Quantile and Bucket Analysis

```python
frame = pd.DataFrame({
    'data1': np.random.randn(1000),
    'data2': np.random.randn(1000)
})

frame
```

```python
quartiles = pd.cut(frame['data1'], 4)
quartiles
```

It's possible to use `Categories` returned by `cut()` in groupy clause

```python
def get_stats(group):
    return {
        'min': group.min(),
        'max': group.max(),
        'count': group.count(),
        'mean': group.mean()
    }
```

```python
grouped = frame['data2'].groupby(quartiles)
grouped.apply(get_stats).unstack()
```

```python
# return quantile numbers
grouping = pd.qcut(frame['data1'], 10, labels=False)
grouped = frame['data2'].groupby(grouping)
grouped.apply(get_stats).unstack()
```

Friendly reminder that `cut` gives equal lenghts buckets, and `qcut` gives us equal-size buckets based on sample quantiles

#### Example: Filling missing values with Group Specific Values

```python
s = pd.Series(np.random.randn(6))

s[::2] = np.nan
s
```

```python
states = ['Ohio', 'New York', 'Vermont', 'Florida', 'Oregon', 'Nevada', 'California', 'Idaho']
group_key = ['East'] * 4 + ['West'] * 4
```

```python
data = pd.Series(np.random.randn(8), index=states)
data[['Vermont', 'Nevada', 'Idaho']] = np.nan
data
```

```python
data.groupby(group_key, group_keys=False).apply(lambda x: x.fillna(x.mean()))
```

```python
fill_val = { 'East': 0.5, 'West': -1 }
data.groupby(group_key, group_keys=False).apply(lambda x: x.fillna(fill_val[x.name]))
```

#### Example: Random Sampliing and Permutations

```python
# Hearts, Spades, Clubs, Diamonds
suits = ['H', 'S', 'C', 'D']
card_val = (list(range(1, 11)) + [10] * 3) * 4
base_names = ['A'] + list(range(2, 11)) + ['J', 'K', 'Q']

cards = []
for suit in ['H', 'S', 'C', 'D']:
    cards.extend(str(num) + suit for num in base_names)

deck = pd.Series(card_val, index=cards)
deck[:13]
```

```python
def draw(deck, n=5):
    return deck.sample(n)
```

```python
draw(deck)
```

```python
# Get two random card foreach suit
get_suit = lambda card: card[-1] # last letter is suit
deck.groupby(get_suit).apply(draw, n=2)
```

#### Example: Group Weighted Average and Correlation

```python
df = pd.DataFrame({
    'category': ['a', 'a', 'a', 'a',
                 'b', 'b', 'b', 'b'],
    'data': np.random.randn(8),
    'weights': np.random.randn(8) })

df
```

#### Example: Group-Wise Linear Regression

```python
import statsmodels.api as sm
def regress(data, yvar, xvars):
    Y = data[yvar]
    X = data[xvars]
    X['intercept'] = 1
    # least squares regression on each chunk of data
    result = sm.OLS(Y, X).fit()
    return result.params
```

```python
by_year.apply(regress, 'AAPL', ['SPX'])
```

## Pivot tables and Cross-Tabulation

*def*: A pivot table is a data summarization tool, that aggregates a table of data by one or more keys, arranging the data in a rectangle with some of the group keys along the rows and some along the columns.

```python
tips
```

```python
# automatically compute the mean of the values!!!!
tips.pivot_table(index=['day', 'smoker'])
```

```python
# The above operation can be solved also as follows
tips.groupby(['day', 'smoker']).mean(numeric_only=True)
```

```python
# aggregate only tip_pct and size, group by time and divided by smoker
tips.pivot_table(['tip_pct', 'size'], index=['time', 'day'], columns='smoker')
```

```python
# as before but we can include partial totals
tips.pivot_table(['tip_pct', 'size'], index=['time', 'day'],
                 columns='smoker', margins=True)
```

```python
#  It's also possibile to specify a different aggregation or NaN filling
tips.pivot_table('tip_pct', index=['time', 'size', 'smoker'],
                 columns='day', aggfunc='mean', fill_value=0)
```

`pivot_table()` supports the following methods
- `values`: column name or names to aggregate; by default, it aggregates a all numeric cols
- `index`: 
- `columns`
- `aggfunc`
- `fill_value`
- `dropna`
- `margins`
